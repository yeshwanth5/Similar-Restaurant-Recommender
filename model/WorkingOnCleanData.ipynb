```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="my_geocoder")
```

```python
df =pd.read_csv('RemovedDuplicatesAndCombinedColumns.csv')
```

```python
df
```

```python
df.location=df.location.apply(lambda x: x+', Bangalore')
```

```python
df
```

```python
def geocode_lat(location):
    gps = geolocator.geocode(location)
    print(gps.latitude)
    return(gps.latitude)
```

```python
df_u = df.location.unique()
```

```python
gps_lat =dict()
for location in df_u:
    gps_lat[location]=geocode_lat(location)
```

```python
gps_lat
```

```python
def geocode_lon(location):
    gps = geolocator.geocode(location)
    print(gps.longitude)
    return(gps.longitude)
```

```python
gps_lon =dict()
for location in df_u:
    gps_lon[location]=geocode_lon(location)
```

```python
df['lat'] = df.location.apply(lambda x : gps_lat.get(x))
```

```python
df['lon'] = df.location.apply(lambda x : gps_lon.get(x))
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
cv_cuisines = cv.fit_transform(df.cuisines)
```

```python
cv_cuisines
```

```python
cv_cuisines = pd.DataFrame(cv_cuisines.toarray(), columns=cv.get_feature_names_out())
```

```python
df = pd.concat([df,cv_cuisines],axis=1)
df.drop('cuisines',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
```

```python
df_vectorized = pd.DataFrame(mn.fit_transform(df.drop(['name','location'],axis=1)))
```

```python
df_vectorized
```

```python
df_vectorized.to_csv('vectorized_data.csv',index=False)
```

```python
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=11).fit(df_vectorized)
```

```python
import pickle
with open('zomato_NearestNeighbour_model.pkl','wb') as f:
    pickle.dump(nbrs,f)
```
